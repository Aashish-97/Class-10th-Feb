```python
# Install dependencies if not already installed
!pip install pandas matplotlib seaborn textblob --quiet

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Mount Google Drive if loading from Drive (optional)
# from google.colab import drive
# drive.mount('/content/drive')

# Load the dataset (Change path accordingly)
file_path = "/content/amazon_review.csv"  # Update path if needed
df = pd.read_csv(file_path, encoding="latin1")

# Function to compute sentiment polarity
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity  # Range: -1 (negative) to 1 (positive)

# Apply sentiment analysis
df["sentiment"] = df["reviewText"].apply(get_sentiment)

# Display summary statistics
print("\nSentiment Summary Statistics:")
print(df[["overall", "sentiment"]].describe())

# Plot Sentiment Score vs. Ratings
plt.figure(figsize=(10, 5))
sns.boxplot(x="overall", y="sentiment", data=df, palette="coolwarm")
plt.title("Sentiment Score vs. Ratings")
plt.xlabel("Rating (1-5 Stars)")
plt.ylabel("Sentiment Score (Polarity)")
plt.grid()
plt.show()

# Save the sentiment-scored dataset (optional)
df.to_csv("amazon_review_with_sentiment.csv", index=False)
print("\n Sentiment analysis completed! Dataset with sentiment scores saved as 'amazon_review_with_sentiment.csv'.")
```

```python
df["reviewTime"] = pd.to_datetime(df["reviewTime"])

# Aggregate by Month
monthly_reviews = df.groupby(df["reviewTime"].dt.to_period("M")).agg({"overall": "mean", "reviewText": "count"})

# Plot Trends
fig, ax1 = plt.subplots(figsize=(12, 5))

ax1.set_xlabel("Month")
ax1.set_ylabel("Average Rating", color="blue")
ax1.plot(monthly_reviews.index.astype(str), monthly_reviews["overall"], color="blue", marker="o", label="Average Rating")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
ax2.set_ylabel("Number of Reviews", color="red")
ax2.plot(monthly_reviews.index.astype(str), monthly_reviews["reviewText"], color="red", marker="o", label="Number of Reviews")
ax2.tick_params(axis='y', labelcolor="red")

plt.title("Review Trends Over Time")
plt.xticks(rotation=45)
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all reviews into a single text
text = " ".join(str(review) for review in df["reviewText"])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Most Common Words in Reviews")
plt.show()
```
