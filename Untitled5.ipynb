```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Upload ORDERS.csv to Colab and adjust the path if needed)
file_path = "/content/ORDERS.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Convert ORDERDATE to datetime format
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors='coerce')

# Extract Year and Month for trend analysis
df["Year"] = df["ORDERDATE"].dt.year
df["Month"] = df["ORDERDATE"].dt.month

# Check for missing values
print("Missing values in the dataset:")
print(df.isnull().sum())

# Drop rows with missing values in important columns
df.dropna(subset=["ORDERDATE", "SALES", "PROFIT"], inplace=True)

# --- 1. Total Sales, Profit, and Order Volume ---
total_sales = df["SALES"].sum()
total_profit = df["PROFIT"].sum()
total_orders = df["ORDERID"].nunique()

print(f"Total Sales: ₹{total_sales:,.2f}")
print(f"Total Profit: ₹{total_profit:,.2f}")
print(f"Total Unique Orders: {total_orders}")

# --- 2. Monthly Sales Trend Analysis ---
monthly_sales = df.groupby(["Year", "Month"])["SALES"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales.index, y=monthly_sales["SALES"], marker="o", label="Monthly Sales")
plt.xlabel("Month (Chronological Order)")
plt.ylabel("Sales Amount (₹)")
plt.title("Monthly Sales Trend")
plt.legend()
plt.show()

# --- 3. Yearly Sales and Profit Trend ---
yearly_sales = df.groupby("Year")[["SALES", "PROFIT"]].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="Year", y="SALES", data=yearly_sales, color="blue", label="Sales")
sns.lineplot(x="Year", y="PROFIT", data=yearly_sales, color="red", marker="o", label="Profit")
plt.xlabel("Year")
plt.ylabel("Amount (₹)")
plt.title("Yearly Sales & Profit Trend")
plt.legend()
plt.show()

# --- 4. Sales Distribution by Order Priority ---
plt.figure(figsize=(10, 5))
sns.boxplot(x="ORDERPRIOTORITY", y="SALES", data=df)
plt.ylim(0, df["SALES"].quantile(0.95))  # Remove extreme outliers
plt.xlabel("Order Priority")
plt.ylabel("Sales Amount (₹)")
plt.title("Sales Distribution by Order Priority")
plt.show()

# --- 5. Top 10 Best-Selling Products ---
top_products = df.groupby("PRODUCTNAME")["SALES"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
top_products.plot(kind="bar", color="green")
plt.xlabel("Product Name")
plt.ylabel("Total Sales (₹)")
plt.title("Top 10 Best-Selling Products")
plt.xticks(rotation=45, ha='right')
plt.show()

# --- 6. Profitability of Product Categories ---
category_profit = df.groupby("PRODUCTCATEGORY")[["SALES", "PROFIT"]].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="PRODUCTCATEGORY", y="PROFIT", data=category_profit, palette="coolwarm")
plt.xlabel("Product Category")
plt.ylabel("Total Profit (₹)")
plt.title("Profitability of Product Categories")
plt.show()

# --- 7. Sales by Customer Segment ---
segment_sales = df.groupby("CUSTOMERSEGMENT")["SALES"].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x="CUSTOMERSEGMENT", y="SALES", data=segment_sales, palette="viridis")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales (₹)")
plt.title("Sales by Customer Segment")
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Upload ORDERS.csv to Colab and adjust the path if needed)
file_path = "/content/ORDERS.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Convert ORDERDATE to datetime format
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors='coerce')

# Check for missing values in key columns
df.dropna(subset=["CUSTOMERSEGMENT", "SALES", "PROFIT", "ORDERQUANTITY"], inplace=True)

# --- 1. Sales and Profitability by Customer Segment ---
segment_performance = df.groupby("CUSTOMERSEGMENT")[["SALES", "PROFIT"]].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="CUSTOMERSEGMENT", y="SALES", data=segment_performance, palette="Blues", label="Total Sales")
plt.xlabel("Customer Segment")
plt.ylabel("Total Sales (₹)")
plt.title("Total Sales by Customer Segment")
plt.xticks(rotation=45)
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x="CUSTOMERSEGMENT", y="PROFIT", data=segment_performance, palette="Reds", label="Total Profit")
plt.xlabel("Customer Segment")
plt.ylabel("Total Profit (₹)")
plt.title("Total Profit by Customer Segment")
plt.xticks(rotation=45)
plt.legend()
plt.show()

# --- 2. Average Order Value (AOV) per Segment ---
df["AOV"] = df["SALES"] / df["ORDERQUANTITY"]
segment_aov = df.groupby("CUSTOMERSEGMENT")["AOV"].mean().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(x="CUSTOMERSEGMENT", y="AOV", data=segment_aov, palette="coolwarm")
plt.xlabel("Customer Segment")
plt.ylabel("Average Order Value (₹)")
plt.title("Average Order Value (AOV) by Customer Segment")
plt.xticks(rotation=45)
plt.show()

# --- 3. Order Priority Distribution by Customer Segment ---
priority_segment = df.pivot_table(index="ORDERPRIOTORITY", columns="CUSTOMERSEGMENT", values="ORDERID", aggfunc="count")

plt.figure(figsize=(10, 6))
sns.heatmap(priority_segment, cmap="YlGnBu", annot=True, fmt=".0f")
plt.xlabel("Customer Segment")
plt.ylabel("Order Priority")
plt.title("Order Priority Distribution by Customer Segment")
plt.show()

# --- 4. Sales Contribution by Customer Segment ---
segment_sales_share = df.groupby("CUSTOMERSEGMENT")["SALES"].sum()
plt.figure(figsize=(8, 8))
segment_sales_share.plot(kind="pie", autopct="%1.1f%%", colors=["lightblue", "lightcoral", "lightgreen", "gold"])
plt.title("Sales Contribution by Customer Segment")
plt.ylabel("")  # Hide y-label
plt.show()

# --- 5. Profitability vs Order Quantity Analysis ---
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["ORDERQUANTITY"], y=df["PROFIT"], hue=df["CUSTOMERSEGMENT"], alpha=0.7, palette="viridis")
plt.xlabel("Order Quantity")
plt.ylabel("Profit (₹)")
plt.title("Profitability vs Order Quantity Across Segments")
plt.legend(title="Customer Segment")
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Ensure ORDERS.csv is uploaded to Colab)
file_path = "/content/ORDERS.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Convert ORDERDATE to datetime format
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors='coerce')

# Drop missing values in key columns
df.dropna(subset=["PRODUCTNAME", "SALES", "PROFIT", "PRODUCTCATEGORY", "DISCOUNT"], inplace=True)

# --- 1. Top 10 Best-Selling Products ---
top_products = df.groupby("PRODUCTNAME")["SALES"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
top_products.plot(kind="bar", color="green")
plt.xlabel("Product Name")
plt.ylabel("Total Sales (₹)")
plt.title("Top 10 Best-Selling Products")
plt.xticks(rotation=45, ha='right')
plt.show()

# --- 2. Most Profitable Products ---
top_profit_products = df.groupby("PRODUCTNAME")["PROFIT"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
top_profit_products.plot(kind="bar", color="red")
plt.xlabel("Product Name")
plt.ylabel("Total Profit (₹)")
plt.title("Top 10 Most Profitable Products")
plt.xticks(rotation=45, ha='right')
plt.show()

# --- 3. Sales & Profit by Product Category ---
category_performance = df.groupby("PRODUCTCATEGORY")[["SALES", "PROFIT"]].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="PRODUCTCATEGORY", y="SALES", data=category_performance, palette="Blues", label="Total Sales")
plt.xlabel("Product Category")
plt.ylabel("Total Sales (₹)")
plt.title("Total Sales by Product Category")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x="PRODUCTCATEGORY", y="PROFIT", data=category_performance, palette="Reds", label="Total Profit")
plt.xlabel("Product Category")
plt.ylabel("Total Profit (₹)")
plt.title("Total Profit by Product Category")
plt.legend()
plt.show()

# --- 4. Discount Impact on Sales & Profit ---
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["DISCOUNT"], y=df["SALES"], alpha=0.5, hue=df["PRODUCTCATEGORY"], palette="coolwarm")
plt.xlabel("Discount (%)")
plt.ylabel("Sales (₹)")
plt.title("Impact of Discounts on Sales")
plt.legend(title="Product Category")
plt.show()

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["DISCOUNT"], y=df["PROFIT"], alpha=0.5, hue=df["PRODUCTCATEGORY"], palette="viridis")
plt.xlabel("Discount (%)")
plt.ylabel("Profit (₹)")
plt.title("Impact of Discounts on Profit")
plt.legend(title="Product Category")
plt.show()

# --- 5. Profit Margins by Product Subcategory ---
df["Profit Margin"] = (df["PROFIT"] / df["SALES"]) * 100
subcategory_margins = df.groupby("PRODUCTSUBCATEGORY")["Profit Margin"].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 5))
subcategory_margins.plot(kind="bar", color="purple")
plt.xlabel("Product Subcategory")
plt.ylabel("Average Profit Margin (%)")
plt.title("Top 10 Product Subcategories with Highest Profit Margins")
plt.xticks(rotation=45, ha='right')
plt.show()

# --- 6. Heatmap of Sales by Product Category and Subcategory ---
pivot_table = df.pivot_table(index="PRODUCTCATEGORY", columns="PRODUCTSUBCATEGORY", values="SALES", aggfunc="sum")

plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, fmt=".0f")
plt.xlabel("Product Subcategory")
plt.ylabel("Product Category")
plt.title("Sales Heatmap: Product Category vs. Subcategory")
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.api as sm

# Load the dataset (Ensure ORDERS.csv is uploaded to Colab)
file_path = "/content/ORDERS.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Drop missing values in key columns
df.dropna(subset=["DISCOUNT", "SALES", "PROFIT"], inplace=True)

# --- 1. Summary Statistics ---
print("Summary of Discounts and Profit:")
print(df[['DISCOUNT', 'PROFIT']].describe())

# --- 2. Scatter Plot: Discount vs. Profit ---
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["DISCOUNT"], y=df["PROFIT"], alpha=0.6, color="blue")
plt.axhline(0, color='red', linestyle='--')  # Mark break-even line
plt.xlabel("Discount (%)")
plt.ylabel("Profit (₹)")
plt.title("Impact of Discounts on Profit")
plt.show()

# --- 3. Discount Ranges & Profitability Analysis ---
df["Discount Range"] = pd.cut(df["DISCOUNT"], bins=[0, 0.05, 0.1, 0.2, 0.3, 0.5, 1], labels=["0-5%", "5-10%", "10-20%", "20-30%", "30-50%", "50%+"])
discount_profit = df.groupby("Discount Range")["PROFIT"].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="Discount Range", y="PROFIT", data=discount_profit, palette="coolwarm")
plt.xlabel("Discount Range")
plt.ylabel("Average Profit (₹)")
plt.title("Average Profit at Different Discount Levels")
plt.show()

# --- 4. Box Plot: Discount vs. Profitability Distribution ---
plt.figure(figsize=(10, 5))
sns.boxplot(x="Discount Range", y="PROFIT", data=df, palette="Set2")
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Discount Range")
plt.ylabel("Profit (₹)")
plt.title("Distribution of Profitability Across Discount Ranges")
plt.show()

# --- 5. Regression Analysis: Discount Impact on Profit ---
df["DISCOUNT_SQ"] = df["DISCOUNT"] ** 2  # Adding squared term for non-linear effect
X = sm.add_constant(df[["DISCOUNT", "DISCOUNT_SQ"]])  # Independent variables
y = df["PROFIT"]  # Dependent variable

model = sm.OLS(y, X).fit()
print(model.summary())

# Generate predictions for visualization
discount_range = np.linspace(0, df["DISCOUNT"].max(), 100)
pred_profit = model.params[0] + model.params[1] * discount_range + model.params[2] * discount_range**2

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["DISCOUNT"], y=df["PROFIT"], alpha=0.4, label="Actual Data")
plt.plot(discount_range, pred_profit, color="red", label="Fitted Curve")
plt.axhline(0, color='black', linestyle='--')  # Break-even line
plt.xlabel("Discount (%)")
plt.ylabel("Predicted Profit (₹)")
plt.title("Regression Model: Impact of Discounts on Profit")
plt.legend()
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Ensure ORDERS.csv is uploaded to Colab)
file_path = "/content/ORDERS.csv"
df = pd.read_csv(file_path, encoding="ISO-8859-1")

# Convert dates to datetime format
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors='coerce')
df["SHIPDATE"] = pd.to_datetime(df["SHIPDATE"], errors='coerce')

# Drop missing values in key columns
df.dropna(subset=["ORDERPRIOTORITY", "SALES", "PROFIT", "SHIPMODE", "SHIPDATE"], inplace=True)

# Create a new column for Shipping Delay (Ship Date - Order Date)
df["Shipping Time (Days)"] = (df["SHIPDATE"] - df["ORDERDATE"]).dt.days

# --- 1. Sales & Profit by Order Priority ---
priority_performance = df.groupby("ORDERPRIOTORITY")[["SALES", "PROFIT"]].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="ORDERPRIOTORITY", y="SALES", data=priority_performance, palette="Blues", label="Total Sales")
plt.xlabel("Order Priority")
plt.ylabel("Total Sales (₹)")
plt.title("Total Sales by Order Priority")
plt.legend()
plt.show()

plt.figure(figsize=(10, 5))
sns.barplot(x="ORDERPRIOTORITY", y="PROFIT", data=priority_performance, palette="Reds", label="Total Profit")
plt.xlabel("Order Priority")
plt.ylabel("Total Profit (₹)")
plt.title("Total Profit by Order Priority")
plt.legend()
plt.show()

# --- 2. Average Order Processing Time per Priority ---
priority_shipping_time = df.groupby("ORDERPRIOTORITY")["Shipping Time (Days)"].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x="ORDERPRIOTORITY", y="Shipping Time (Days)", data=priority_shipping_time, palette="coolwarm")
plt.xlabel("Order Priority")
plt.ylabel("Average Shipping Time (Days)")
plt.title("Average Order Processing Time by Order Priority")
plt.show()

# --- 3. Profitability Distribution Across Priorities ---
plt.figure(figsize=(10, 5))
sns.boxplot(x="ORDERPRIOTORITY", y="PROFIT", data=df, palette="Set2")
plt.axhline(0, color='red', linestyle='--')
plt.xlabel("Order Priority")
plt.ylabel("Profit (₹)")
plt.title("Distribution of Profitability Across Order Priorities")
plt.show()

# --- 4. Order Priority vs. Shipping Mode ---
priority_shipping_mode = df.pivot_table(index="ORDERPRIOTORITY", columns="SHIPMODE", values="ORDERID", aggfunc="count")

plt.figure(figsize=(10, 6))
sns.heatmap(priority_shipping_mode, cmap="YlGnBu", annot=True, fmt=".0f")
plt.xlabel("Shipping Mode")
plt.ylabel("Order Priority")
plt.title("Order Priority vs. Shipping Mode")
plt.show()
```
